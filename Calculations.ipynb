```python
import pandas
import glob
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv('/home/ozgur/workspace/ns-allinone-3.30.1/ns-3.30.1/xml/VanetRC-report.csv', header=None)
df2 = pandas.read_csv('/home/ozgur/workspace/ns-allinone-3.30.1/ns-3.30.1/xml/MeshTest-report.csv', header=None)
connections = [5, 10, 15, 20]
pdfOurs = []
RXbitrateOurs = []
pdfTheirs = []
RXbitrateTheirs = []
'''
pdfOurs = [90.28, 97.18, 98.54, 98.11]
RXbitrateOurs = [756.09, 1374.43, 1967.31, 2699.27]

pdfTheirs = [98.41, 98.23, 99.26, 89.54]
RXbitrateTheirs = [737.34, 1473.81, 2203.75,  2671.18]
'''
for x in range(1,5):
    tmp = df.loc[df[0] == 5*x]
    pdfOurs.append(tmp[2].mean())
    RXbitrateOurs.append(tmp[3].mean())
    
    tmp = df2.loc[df2[0] == 5*x]
    pdfTheirs.append(tmp[2].mean())
    RXbitrateTheirs.append(tmp[3].mean())
```

```python
plt.plot(connections, pdfOurs, marker='o', markerfacecolor='blue', label="Bizinkisi")
plt.plot(connections, pdfTheirs, marker='x', markerfacecolor='red', label="Onlarınkisi")
plt.legend(loc='lower left')
plt.ylabel('Packet Delivery Fraction (%)')
plt.xlabel('Connections')
plt.xticks([5, 10, 15, 20])
```

```python
plt.plot(connections, RXbitrateOurs, marker='o', markerfacecolor='blue', label="Bizinkisi")
plt.plot(connections, RXbitrateTheirs, marker='x', markerfacecolor='red', label="Onlarınkisi")
plt.legend(loc='upper left')
plt.ylabel('RXbitrate (kbps)')
plt.xlabel('Connections')
plt.xticks([5, 10, 15, 20])
```
